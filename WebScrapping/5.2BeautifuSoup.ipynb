## Putting it all together

```python
title =[]
price = []
availibility =[]
review = []
link = []
star = []

def books(soup):
    sp_prices = soup.find_all('p', class_='price green')
    sp_titles = soup.find_all('p', class_='book_name')
    sp_stocks = soup.find_all("p", class_='stock')
    sp_reviews = soup.find_all("p", class_='review')
    
    bks = soup.find_all('div', class_ = 'book_container')
    for bk in bks:
        star.append(5 - (len(bk.find_all("span", class_='not_filled'))))    
    
    for i in range(len(sp_titles)):
        title.append(sp_titles[i].text)
        price.append(sp_prices[i].text)
        availibility.append(sp_stocks[i].text)
        review.append(sp_reviews[i].get("data-rating"))
        link.append(sp_titles[i].find('a').get('href'))
        
req = requests.get("https://arifpucit.github.io/bss2/index.html")
soup = BeautifulSoup(req.text, 'lxml')
books(soup)

data = {
    "Title/Author":title,
    "Price": price,
    "Avalibility": availibility,
    "Rating":star,
    "Reviews":review,
    "Link":link
}
df = pd.DataFrame(data=data, columns=["Title/Author", "Price", "Avalibility","Rating", "Reviews","Link"])
df.to_csv("books3.csv",index=False)
df = pd.read_csv("books3.csv")
df
```

### Get Content from multiple Webpages at the same time

```python
title =[]
price = []
availibility =[]
review = []
link = []
star = []

def books(soup):
    sp_prices = soup.find_all('p', class_='price green')
    sp_titles = soup.find_all('p', class_='book_name')
    sp_stocks = soup.find_all("p", class_='stock')
    sp_reviews = soup.find_all("p", class_='review')
    
    bks = soup.find_all('div', class_ = 'book_container')
    for bk in bks:
        star.append(5 - (len(bk.find_all("span", class_='not_filled'))))    
    
    for i in range(len(sp_titles)):
        title.append(sp_titles[i].text)
        price.append(sp_prices[i].text)
        availibility.append(sp_stocks[i].text)
        review.append(sp_reviews[i].get("data-rating"))
        link.append(sp_titles[i].find('a').get('href'))
        
urls = ['https://arifpucit.github.io/bss2/index.html',
        'https://arifpucit.github.io/bss2/SP.html',
        'https://arifpucit.github.io/bss2/CA.html']
for url in urls:
    req = requests.get(url)
    soup = BeautifulSoup(req.text, 'lxml')
    books(soup)

data = {
    "Title/Author":title,
    "Price": price,
    "Avalibility": availibility,
    "Rating":star,
    "Reviews":review,
    "Link":link
}
df = pd.DataFrame(data=data, columns=["Title/Author", "Price", "Avalibility","Rating", "Reviews","Link"])
df.to_csv("books3.csv",index=False)
df = pd.read_csv("books3.csv")
df
```

### Getting Data from all Different Navigation Number Pages

```python
def Videos(soup):
    articles = soup.find_all('div', class_ = 'media-body')
    for article in articles:
        title = article.find('h4', class_= 'media-heading1').text
        titles.append(title)
        
        desc = article.find('p', align = 'justify').text
        description.append(desc)
        
        video_id = article.find('iframe')['src'].split('/')[4].split('?')[0]
        youtube_id = f'https://youtube.com/watch?v={video_id}'
        links.append(youtube_id)
```

```python
titles = []
description = []
links = []

first = requests.get("http://www.arifbutt.me/category/sp-with-linux/")
soup = BeautifulSoup(first.text, 'lxml')
Videos(soup)

while True:
    pagination_code = soup.find('div', class_ = 'navigation_pegination')
    links_all = pagination_code.find_all('li')
    lastlink = None
    for lastlink in links_all:
        pass
    if(lastlink.find('a').text == "Next Page »"):
        next_url= lastlink.find('a').get('href')
        req = requests.get(next_url)
        soup = BeautifulSoup(req.text, 'lxml')
        Videos(soup)
    else:
        break;

data = {
    "Title":titles,
    "Youtube Link": links,
    "Description": description
}
df = pd.DataFrame(data=data, columns=["Title", "Youtube Link", "Description"])
df.to_csv("AllVideos.csv",index=False)
df = pd.read_csv("AllVideos.csv")
df
```
