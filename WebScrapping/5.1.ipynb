## 2. request library
### Provides cookies and sessions support which was not provided in urllib

```python
import requests
```

```python
res2 = requests.get("https://arifpucit.github.io/bss2/index.html")
res2
```

```python
res2.status_code
```

```python
res2.url
```

```python
res2.headers
```

```python
res2.headers['Server']
```

```python
res2.headers['Last-Modified']
```

```python
res2.text
```

### Save content in txt file

```python
with open('bss2-index.html', 'wt') as fd:
    fd.write(res2.text)
```

### Save image in a binary file

```python
url ="https://arifpucit.github.io/bss2/" + "images/OS concepts.jpg"
reqI = requests.get(url)
reqI
```

```python
reqI.content  # FOR IMAGES USE CONTENT PARAM
```

```python
with open('Myimg.png', 'wb') as fd:
    fd.write(reqI.content)
```
