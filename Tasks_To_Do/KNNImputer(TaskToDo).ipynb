## KNN Imputer (Task To Do)...
### Load Dataset

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
df =  pd.read_csv("datasets/loan-eligibility.csv")
df.head()
```

```python
df.info()
```

### Step1 is to Do Ctegorical Encoding as KNN Imputer works on Numerical Data

```python
X = df.drop("Loan_Status", axis=1)
X.head()
```

### Label Encoding

```python
lb = LabelEncoder()
df['Loan_Status'] = lb.fit_transform(df['Loan_Status'])
df.head()
```

### Ordinal Encoding On Property_Area

```python
df.Property_Area.value_counts()
```

```python
df.Dependents.value_counts()
```

```python
df_property=df.iloc[:,11:12]
oe = OrdinalEncoder(categories=[['Semiurban', 'Urban', 'Rural']],dtype=np.int32)
oe.fit(df_property)
oe.categories_
```

```python
df['Property_Area'] = oe.transform(df_property)
df.head()
```

```python
df_dependents=df.iloc[:,3:4]
oe = OrdinalEncoder(categories=[['0', '1', '2', '3+']],dtype=np.float32, handle_unknown='use_encoded_value', unknown_value=np.nan)
oe.fit(df_dependents)
oe.categories_
```

```python
df['Dependents'] = oe.transform(df_dependents)
df.head()
```

### Nominal Enconding on Rest of Categorical Coulmns

```python
df.Education.value_counts()
```

```python
df.Gender.value_counts()
```

```python
df.Self_Employed.value_counts()
```

```python
df.Married.value_counts()
```

```python
df_education = df.iloc[:,4:5]
```

```python
ohe = OneHotEncoder(drop='first', sparse=False, dtype=np.int32)
ohe.fit(df_education)
ohe.categories_
```

```python
ohe.transform(df_education)
df['Education'] = ohe.transform(df_education)
df.rename(mapper={'Graduate':'Not_Graduated'}, axis=1, inplace=True)
df.head()
```

```python
df['Gender'].replace({'Male':0, 'Female':1}, inplace=True)
df['Married'].replace({'No':0, 'Yes':1}, inplace=True)
df['Self_Employed'].replace({'No':0, 'Yes':1}, inplace=True)
df.rename(mapper={'Gender':'Female'}, axis=1, inplace=True)
```

```python
df
```

```python
df.info()
```

### Imputer

```python
X = df.drop("Loan_ID", axis=1)
y = df['Loan_ID']
imp = KNNImputer(missing_values=np.nan, n_neighbors=2, weights='uniform', metric='nan_euclidean')
imp.fit(X)
```

```python
X = imp.transform(X)
X
```

```python
col = ['Female','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome',
       'LoanAmount', 'Loan_Amount_Term','Credit_History', 'Property_Area', 'Loan_Status']
df_imputed2 = pd.DataFrame(data=X, columns=col)
df_imputed2
```

```python
df_imputed1 = pd.DataFrame(data=y, columns=['Loan_ID'])
df_imputed1
```

```python
res_df = result = pd.concat([df_imputed1, df_imputed2], axis=1)
res_df.info()
```
